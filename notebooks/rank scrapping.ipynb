2022 LEAGUE RANKING

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display, HTML

# Send a GET request to the URL
url = 'https://www.fifaindex.com/teams/fifa23_580/?league=13&order=desc'
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the team data
table = soup.find('table', class_='table-teams')

# Extract the data from the table
team_data = []
header_row = table.find('thead').find('tr')
for row in table.find('tbody').find_all('tr'):
    data_row = {}
    for idx, cell in enumerate(row.find_all('td')):
        header = header_row.find_all('th')[idx].text.strip()
        value = cell.text.strip()
        data_row[header] = value
    team_data.append(data_row)

# Create a pandas DataFrame from the extracted data
df = pd.DataFrame(team_data)

# Select and reorder the desired columns
columns = ['Name', 'ATT', 'MID', 'DEF', 'OVR']

# Add the 'League Year' column with the value '2022'
df['League Year'] = '2022'

# Reorder the columns to include 'League Year' as the first column
df = df[['League Year'] + columns]

# Drop rows with NaN values
df = df.dropna()

# Reset the index
df = df.reset_index(drop=True)

df = df.rename(columns={'Name': 'Team Name', 'ATT': 'Attack', 'MID': 'Midfield', 'DEF': 'Defence', 'OVR': 'Overall'})

# Set the index to start from 1
df.index = df.index + 1

# Format the DataFrame to display with lines
styles = [
    dict(selector="th", props=[("border", "1px solid black")]),
    dict(selector="td", props=[("border", "1px solid black")]),
]

styled_df = df.style.set_table_styles(styles)

# Display the styled DataFrame
display(HTML(styled_df.render()))
```

```python
# Save the DataFrame as a CSV file
df.to_csv('2022_team_ranking_data.csv', index=False)
```
